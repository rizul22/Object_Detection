```python
import torch

# Load YOLOv5 pre-trained model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')  # 'yolov5s' is the smallest version

# Check model details (optional)
model.eval()  # Set model to evaluation mode
```

```python
import cv2
import matplotlib.pyplot as plt

# Load and display an image
def load_image(image_path):
    img = cv2.imread(image_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB
    plt.imshow(img_rgb)
    plt.axis('off')  # Hide axes
    plt.show()
    return img

# Example image path
image_path = r"./images/download.png"
img = load_image(image_path)
```

```python
# Perform object detection
results = model(img)

# Display the image with detected objects
results.show()
```

```python
# Get the predictions (bounding boxes, confidence scores, and class labels)
detections = results.pandas().xyxy[0]  # This returns a pandas DataFrame

# Count the number of objects detected
num_objects = len(detections)
print(f"Number of objects detected: {num_objects}")
```

```python
# Filter detections for a specific object class (e.g., 'person')
person_detections = detections[detections['name'] == 'person']

# Count the number of people detected
num_people = len(person_detections)
print(f"Number of people detected: {num_people}")
```
